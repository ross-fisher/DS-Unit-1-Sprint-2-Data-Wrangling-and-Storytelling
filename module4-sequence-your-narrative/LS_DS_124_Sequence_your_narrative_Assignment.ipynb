```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
income = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--income_per_person_gdppercapita_ppp_inflation_adjusted--by--geo--time.csv')
lifespan = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--life_expectancy_years--by--geo--time.csv')
population = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--population_total--by--geo--time.csv')
entities = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')
concepts = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--concepts.csv')
```

```python
# inner join, merges only on rows where 
# the same columns match in values
df = pd.merge(income, lifespan)
print(df.shape)
df = pd.merge(income, lifespan, on=['geo','time'], how='inner')
print(df.shape)

df.head()
```

```python
# population : geo,time,population_total
df = pd.merge(df, population) 
print(df.shape)
df.head()
```

```python
columns = ['country', 'name', 'world_4region', 'world_6region']
subset = entities[columns]
subset.head()
```

```python
merged = pd.merge(df, subset, left_on='geo', right_on='country')
merged.head()
```

```python
merged = merged.drop('geo', axis='columns')
merged.head()
```

```python
merged = merged.rename(columns = {
    'country': 'country_code',
    'time': 'year', 
    'income_per_person_gdppercapita_ppp_inflation_adjusted': 'income', 
    'life_expectancy_years': 'lifespan',
    'population_total': 'population',
    'name': 'country',
    'world_6region': '6region',
    'world_4region': '4region'
})

merged.head()
```

```python
import matplotlib
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

colormap = {'south_asia' : 'blue',
            'sub_saharan_africa':'orange',
            'europe_central_asia':'green',
            'middle_east_north_africa':'maroon',
            'america':'cyan',
            'east_asia_pacific':'brown'
           }


patches = [mpatches.Patch(color=c,label=l) for l,c in zip(colormap.keys(),colormap.values())]
end_game = merged[merged.year==2018][merged.lifespan < 60]

for i,year in enumerate(years):
    df = merged[merged.year == year]
    fig = plt.figure(figsize=(10,5))
    if i == 0: 
        plt.suptitle('Countries with lifespan < 60',x=.2,y=1.15,fontsize=20,fontweight='bold')
    # [left bottom width height]
    ax  = fig.add_axes([0.0, 0.0, 0.66, 1.0])
    ax.set_title(year)
    ax.set_xlim(0,90)
    ax.set_xlabel('Lifespan')
    ax.set_ylabel('Income')
    plt.style.use("fivethirtyeight")
    sub = df[df.lifespan < 60]
    colorlist = sub['6region'].apply(lambda x: colormap[x])
    ax.scatter(sub.lifespan,sub.income,color=colorlist)
    
    
    plt.legend(handles=patches, loc='upper left',fontsize=10)

# plot the names of every country left at the end
end_guys = pd.merge(end_game, sub)
for country in end_guys.country:
    sub = merged[merged.year==2018][merged.lifespan<60][merged.country == country].iloc[0]
    ax.text(x=sub.lifespan, y=sub.income,s=sub.country)
```
