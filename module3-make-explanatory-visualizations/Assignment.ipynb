```python
#https://fivethirtyeight.com/features/the-ultimate-halloween-candy-power-ranking/
candy = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/candy-power-ranking/candy-data.csv')
candy.head()
```

```python
candy_by_rank = candy.sort_values(by='winpercent',ascending=False)
rank_listing = list(candy_by_rank.competitorname)
df = pd.DataFrame({'RK'    : range(1,candy.shape[0]+1),
              'Candy' : candy_by_rank.competitorname,
              'Percentage' : candy_by_rank.winpercent})
df = df.set_index('RK')
df.head()             
```

```python
def color_negative_red(value):
  """
  Colors elements in a dateframe
  green if positive and red if
  negative. Does not color NaN
  values.
  """

#   if value < 50:
#     color = 'red'
#   elif value > 0:
#     color = 'green'
#   else:
#     color = 'black'

#   return 'color: %s' % color
  return "color: black;background-color: white"
```

```python
# .format({'total_amt_usd_pct_diff': "{:.2%}"}))
df.style.applymap(color_negative_red, subset=['Percentage'])
x = (df
  .head()
  .style
  .bar(subset=['Percentage'], color='#ED713A')#, color='#red')
  .applymap(color_negative_red, subset=['Percentage'])
  .format({'Percentage': '{:.3}%'})
)
x
```

```python
fig,ax = plt.subplots(figsize=(9, 17))
plt.style.use("fivethirtyeight")
subset = df.head(85)
import matplotlib
from matplotlib.patches import Rectangle

plt.xlim(0,100)
plt.ylim(0,86)
ax.invert_yaxis()

ax.text(0, 0, 'RK', fontsize=10, fontweight='bold')
ax.text(7, 0, 'Candy', fontsize=10, fontweight='bold')
ax.text(60, 0, 'Win Percentage', fontsize=10, fontweight='bold')

ax.plot([0,100],[.5,.5], color='black', linewidth=0.8)

ax.get_yaxis().set_visible(False)
ax.get_xaxis().set_visible(False)

# draw everything from the dataframe
for y in range(1,85):
  ax.text(2, y+.5, y, fontsize=11)
  ax.text(7, y+.5, df.iloc[y].Candy, fontsize=11)
  percent = df.iloc[y].Percentage
  ax.text(60,y+.5, '{:.3}%'.format(percent), fontsize=11)
  # y-1 becasue array is from 0
  rect = Rectangle((70,y-.3), 30*(percent/100),0.7,color='#ED713A')
  ax.add_patch(rect)
  ax.plot([0,100], [y + 0.7, y+0.7], color='black', linewidth=0.4)
```
